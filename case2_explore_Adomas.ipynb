```python
import pandas as pd
import plotly.express as px

```

```python
df = pd.read_csv('HR_data.csv').drop(columns=['Unnamed: 0'])
df.head()
```

```python
pd.set_option('display.max_rows', None)
df.describe().T
```

```python
pd.reset_option('display.max_rows')
```

```python
# separate feature groups
emotion_cols = [ 'Frustrated', 'upset', 'hostile', 'alert',
       'ashamed', 'inspired', 'nervous', 'attentive', 'afraid', 'active',
       'determined']

feature_cols = [col for col in df.columns if '_' in col]

cat_cols = [col for col in df.columns if col not in emotion_cols+feature_cols]
```

```python
df.Cohort.sort_values().unique()
# why the split of cohort d13? 
```

```python
fig = px.scatter(
    data_frame=df[cat_cols].sort_values(by='Cohort'),
    x = 'Cohort',
    y = 'Individual',
    color='Cohort',
    template='simple_white',
    
)
fig.update_layout(
    width = 500,
    title = 'Are cohorts unique?'
)

fig.update_xaxes(showgrid=True, gridcolor='grey')
fig.update_yaxes(showgrid=True, gridcolor='grey',minor=dict(tickcolor="white", showgrid=True))
fig.show()
```

## pca for Cohort and other labels

```python

# Load the data
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import StandardScaler


data = pd.read_csv('HR_data.csv').drop(columns=['Unnamed: 0'])
data = data.dropna()

# Columns to exclude from PCA
cat_cols = [
     'Round', 'Phase', 'Cohort', 'Puzzler', 'Individual'
]
# Columns that are emotional states
emotion_cols = [
    'Frustrated', 'upset', 'hostile', 'alert', 'ashamed', 
    'inspired', 'nervous', 'attentive', 'afraid', 'active', 'determined'
]
feature_cols = [col for col in data.columns if col not in cat_cols+emotion_cols]

# prep the three datasets for PCA
feature_df = data[feature_cols].copy()
emo_df = data[emotion_cols].copy()
combo_df = data[feature_cols+emotion_cols].copy()

# Scaler Instance
scaler = StandardScaler()
#recombine cohort 3
data['cohort_recom'] = data.Cohort.str.split('_', expand=True)[1]
data['cohort_recom'] = data.cohort_recom.astype(int).apply(lambda x: 'd1_3' if x >= 3 else f'd1_{x}')

# choose labeling target
target = 'cohort_recom'


for df, label in zip([feature_df, emo_df, combo_df],['Feature PCA','Emotion PCA', 'PCA Score Plot']):
    df_scaled = scaler.fit_transform(df)

    # Apply PCA
    pca = PCA(n_components=2)  # Adjust based on your analysis
    principal_components = pca.fit_transform(df_scaled)

    

    # Create a DataFrame for plotting
    pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
    pca_df[target] = data[target]#.apply(lambda x: 'yes' if x == 1 else 'no')  
    # Plotting PCA results color-coded by emotion
    fig, ax = plt.subplots(figsize=(6, 4), dpi = 120)
    scatter = sns.scatterplot(x=pca_df['PC1'], y=pca_df['PC2'], hue=pca_df[target],s=20)
    plt.xlabel(f'Principal Component 1 ({round(pca.explained_variance_ratio_[0] * 100, 1)}%)')
    plt.ylabel(f'Principal Component 2 ({round(pca.explained_variance_ratio_[1] * 100, 1)}%)')
    plt.title(f'{label}')
    plt.legend(title = 'Cohort')
    #plt.colorbar(scatter)
    plt.grid(True)
    plt.show()
```

```python
# Assuming 'features_scaled' has already been prepared and is ready for PCA
import numpy as np


pca = PCA()
components = pca.fit_transform(scaler.fit_transform(combo_df))

# Create a DataFrame with the loadings
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(len(pca.components_))], index=combo_df.columns)

# Plotting the explained variance
plt.figure(figsize=(8, 4))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.hlines(y = 0.95, xmax=60, xmin=0, linestyles='dashed')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot of Explained Variance')
plt.grid(True)
plt.show()

# Visualize the loadings for the first few components
plt.figure(figsize=(12, 16), dpi=120)
sns.heatmap(loadings.iloc[:, :2], cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Component Loadings')
plt.show()
```

### Look into emotion score diff phase1 -> phase 2

```python
emo_change_df = cohort_grouped.loc[cohort_grouped.Phase != 'phase3'].copy()
emo_change_df.sort_values(by=['cohort_recom','Phase',], ascending=False)
```

```python
# Assuming your data is stored in a pandas DataFrame called 'emo_change_df'

# Step 1: Filter Data by Phase
phase1_data = emo_change_df[emo_change_df['Phase'] == 'phase1']
phase2_data = emo_change_df[emo_change_df['Phase'] == 'phase2']

# Step 2: Group by Cohort Recommendation
phase1_grouped = phase1_data.groupby('cohort_recom')
phase2_grouped = phase2_data.groupby('cohort_recom')

# Step 3: Calculate Mean Difference for Each Emotion
emotion_diff = {}
for emotion in emo_change_df.columns[2:]:  # Assuming the emotions start from the 4th column
    diff = {}
    for cohort, group1 in phase1_grouped:
        group2 = phase2_grouped.get_group(cohort)
        diff[cohort] = group2[emotion].mean() - group1[emotion].mean()
    emotion_diff[emotion] = diff

# Convert dictionary to DataFrame for easier manipulation
emotion_diff_df = pd.DataFrame(emotion_diff)

# Display the resulting DataFrame
emotion_diff_df
```

```python
emotion_diff_df = emotion_diff_df.reset_index().melt(id_vars='index')
```

```python
fig, ax = plt.subplots(figsize = (8,4), dpi = 120)
sns.barplot(
    data=emotion_diff_df,
    x='variable',
    y = 'value',
    hue = 'index'
)
sns.despine()
plt.title('Mean Emotion Score Change: Phase_1 -> Phase_2')
plt.ylabel('Difference')
plt.xlabel(' ')
plt.legend(title = 'Cohort')
plt.xticks(rotation = 60)
plt.show()
```
